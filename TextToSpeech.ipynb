### Tacotron2-DDC


```python
from TTS.api import TTS

tts = TTS(model_name="tts_models/de/thorsten/tacotron2-DDC", progress_bar=False, gpu=False)

```

```python
from IPython.display import Audio
```

```python

text = "Hello, this is a sample text to speech conversion using Coqui TTS by Niharika"

tts.tts_to_file(text=text, file_path="tacotron-ddc.wav")
```

```python

Audio("tacotron-ddc.wav")
```
